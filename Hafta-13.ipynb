```python
df_cust["fea_2"].fillna(df_cust["fea_2"].mean(),inplace=True) # Boş olan alanları doldurduk
df_cust.isna().sum()
```

```python
y = df_cust["label"]
X = df_cust.drop(["label","id"],axis=1)
X
```

```python
from imblearn.over_sampling import RandomOverSampler #var olan verileri kullanarak yeni veri üretmemizi sağlar
from collections import Counter
```

```python
sampler = RandomOverSampler()
X_new,y_new = sampler.fit_resample(X,y)
print(f"Eski Durum:{Counter(y)}")
print(f"Yeni Durum:{Counter(y_new)}")
```

```python
from sklearn.model_selection import train_test_split #Train ve test olarak bölmemizi sağlar
```

```python
df_cust_new = X_new.join(y_new)
plt.figure(figsize=(14,8))
sns.heatmap(df_cust_new.corr(),annot=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_new,y_new,test_size=0.2,random_state=42) #100 verinin 80 ini train 20 sini test olarak ayırıyor
```

```python
from sklearn.tree import DecisionTreeClassifier #Sınıflandırma yapacağımızı belirtiyoruz
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
model_tree = DecisionTreeClassifier()
model_tree.fit(X_train,y_train) #modelimizi eğittik
```

```python
print(f"Train Accuracy Score = {accuracy_score(y_train,model_tree.predict(X_train))}")
print(f"Test Accuracy Score = {accuracy_score(y_test,model_tree.predict(X_test))}")
#Aşırı öğrenme var hiçbir zaman 1.0 lık bir tahmin olamaz
```

```python
from sklearn import tree
```

```python
tree.plot_tree(model_tree) #karar ağacını veriyor ama ağaç cok dallandığı için onu budayacağız aşırı öğrenme söz konusu
```

```python
# ccp_aplha minimum maliyet-karmaşıklık budaması için kullanılan karmaşıklık parametresi
# En büyük maliyet karmaşıklığına sahip olan ccp_alpha dan küçük olan alt ağaçlar seçilir
# Varsayılan olarak budama yapılmaz
```

```python
path = model_tree.cost_complexity_pruning_path(X_train,y_train)
alphas = path["ccp_alphas"]
path["ccp_alphas"]
```

```python
accuracy_train,accuracy_test=[],[]
for i in alphas:
    model_prune = DecisionTreeClassifier(ccp_alpha=i)
    
    model_prune.fit(X_train,y_train)
    y_train_pred=model_prune.predict(X_train)
    y_test_pred = model_prune.predict(X_test)
    
    accuracy_train.append(accuracy_score(y_train,y_train_pred))
    accuracy_test.append(accuracy_score(y_test,y_test_pred))
```

```python
sns.lineplot(y=accuracy_train,x=alphas,label='Eğitim Doğruluğu')
sns.lineplot(y=accuracy_test,x=alphas,label='Test Doğruluğu')
plt.xlabel("Alfa Değerleri")
plt.ylabel("Doğruluk")
plt.xticks()
plt.show()
```

```python
model_tree_new = DecisionTreeClassifier(ccp_alpha=0.006) # 0.006 prune değeridir paralelleşmeye başladığı an
model_tree_new.fit(X_train,y_train)
```

```python
print(f"Train Accuracy Score = {accuracy_score(y_train,model_tree_new.predict(X_train))}")
print(f"Test Accuracy Score = {accuracy_score(y_test,model_tree_new.predict(X_test))}")
```

```python
tree.plot_tree(model_tree_new)
```

```python
sns.heatmap(confusion_matrix(y_test,model_tree_new.predict(X_test)),annot=True,fmt='g')
#fmt g bize sayısal olarak farklı şekilde gösterir 
#y_testteki gerçek değerler ile y_test modelimizi karşılaştırıyor
```
